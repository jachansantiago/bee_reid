```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("/home/jchan/beeid/notebooks/cmc_experiments/data/test_unnormalized.csv")
```

```python
track_lens = df.groupby("global_track_id").apply(len)
total_tracks = len(track_lens)
total_tracks
```

```python
value_counts = track_lens.value_counts()
cummulative=total_tracks

xs = list()
ys = list()

xs.append(1)
ys.append(cummulative/total_tracks)

for i in range(15):
    cummulative -= value_counts[i+1]
    xs.append(i+2)
    
    ys.append(cummulative/total_tracks)
```

```python
plt.figure(figsize=(20, 10))

plt.plot(xs, ys, "-^", markersize=8)
plt.grid(True)
plt.ylim(0.0, 1.0)
plt.yticks(np.arange(0, 1.10, 0.05), fontsize=16);
plt.xticks(np.arange(1, 15 + 1, 1), fontsize=16);
plt.xlabel("at least x images per tracks", fontsize=22)
plt.ylabel("Fraction of total number of tracks", fontsize=22);
# plt.legend(fontsize=22)
plt.tight_layout()
plt.savefig("track_len_versus_fraction.png", bbox_inches="tight")
```
