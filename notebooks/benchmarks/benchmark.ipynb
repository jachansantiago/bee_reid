```python
paper_benchmark = new_benchmark[new_benchmark.benchmark == "new"].drop([1,8,  12, 14])
# paper_benchmark["training protocol"] = ["long-term", "short-term", "short-term + long-term", "long-term", "short-term + long-term", "short-term"]
paper_benchmark["training protocol"] = ["long-term", "short-term", "short-term + long-term", "long-term", "short-term + long-term", "short-term"]
paper_benchmark["augmentation"] = [False, True, True, True, False, False]

no_aug = paper_benchmark[paper_benchmark.augmentation == False]
aug = paper_benchmark[paper_benchmark.augmentation == True]
```

```python
pb = no_aug.merge(aug, on="training protocol", suffixes=("", "_aug"))
```

```python
pb.columns
```

```python
pb
```

```python
cols = pb.columns.tolist()
cols[7] = "same day; same hour"
cols[9] = "diff day; same hour"
cols[11] = "diff day; any hour"
cols[23] = "same day; same hour w/ aug"
cols[25] = "diff day; same hour w/ aug"
cols[27] = "diff day; any hour w/ aug"
pb.columns = cols
```

```python
# Sort Training Protocols
from pandas.api.types import CategoricalDtype
cat_size_order = CategoricalDtype(
    ['long-term', 'short-term', 'short-term + long-term'], 
    ordered=True
)
pb["training protocol"] = pb["training protocol"].astype(cat_size_order)
pb = pb.sort_values("training protocol")
```

```python
eval_protocols = ["same day; same hour",
                  "diff day; same hour",
                  "diff day; any hour"]
colors = ["tab:blue", "tab:orange", "tab:red"]
patterns = [ "|" , "\\" , "/" ]
fig, ax = plt.subplots(figsize=(10, 10))
pb.plot(x="training protocol", y=eval_protocols, kind="bar", ylabel="CMC @ Rank 1", ax=ax, color=colors)
plt.xticks(rotation=0);

# bars = ax.patches
# patterns =('-', '+', 'x')
# hatches = [p for p in patterns for i in range(3)]
# for bar, hatch in zip(bars, hatches):
#     bar.set_hatch(hatch)
```

```python
nbenchmark = load_benchmark("../../models3/")
nbenchmark["n"] = nbenchmark["date"].str[-4:].astype(int)
```

```python
nbenchmark.sort_values("n")
```

```python
lt_dict = {
    "n": [181, 181, 181],
    "eval": ["rank1_same_hour", "rank1_diff_day_same_hour", "rank1_diff_day"],
    "train": ["long-term", "long-term", "long-term"],
    "value": [0.529, 0.391, 0.362]
}
ltdf = pd.DataFrame(lt_dict)
```

```python
ddff = pd.melt(nbenchmark, value_vars=['rank1_same_hour', 'rank1_diff_day_same_hour', 'rank1_diff_day'], id_vars="n")
ddff["train"] = "short-term"
c = list(ddff.columns.copy())
c[1] = "eval"
ddff.columns = c
ddff = ddff.append(ltdf)
```

```python
num_tracks = [181, 362, 724, 1448, 2896, 4949]

plt.figure(figsize=(13, 8))

p = sns.lineplot(x="n", y="value", style="train", hue="eval", markers=True, dashes=False, markersize=16,
         linewidth=5, palette=['b', 'r', 'g'], data=ddff)



# plt.plot(x, x/11, ':', linewidth=2, color="gray")
plt.grid(True)
plt.ylim(0.0, 1.0)
plt.yticks(np.arange(0, 1.05, 0.05), fontsize=22);

plt.xlabel("# IDs", fontsize=28)
plt.ylabel("Accuracy", fontsize=28)
plt.xscale('log')
plt.xticks(num_tracks, ["1X", "2X", "4X", "8X", "16X", "27.3X"], fontsize=22);
plt.tick_params(axis='y', which='minor', bottom=False)
plt.title("Accuracy versus number of tracks.", fontsize=32);
plt.legend(fontsize=18, ncol=2)
plt.tight_layout()
plt.savefig("performanceversusntracks3.png", bbox_inches="tight")
```
