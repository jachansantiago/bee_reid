### Evaluation of number of Tracks

```python
def get_curve_ntrack(df_dict, column="test_same_hour", rank=1):
    xs = list()
    ys = list()
    for num_track, df in df_dict.items():
        xs.append(num_track)
        y = df[column].values[rank - 1]
        ys.append(y)
    return xs, ys
    
```

```python
num_tracks = [181, 362, 724, 1448, 2896, 4949]
csv_format = "results/pretraining_triplet_loss_{}_ids_full_evaluation.csv"
df_dict = dict()

for nt in num_tracks:
    csv_path = csv_format.format(nt)
    df = pd.read_csv(csv_path)
    df_dict[nt] = df.copy()
    
X = list()
Y = list()
L1 = list()
L2 = list()



xs, ys = [181], [triplet_df["test_same_hour"][0]]

X += xs
Y += ys
L1 += ["same hour rank 1"]
L2 += ["long-term"]


xs, ys = [181], [triplet_df["test_different_day_same_hour"][0]]

X += xs
Y += ys
L1 += ["same hour diff day rank 1"]
L2 += ["long-term"]


xs, ys = [181], [triplet_df["test_different_day"][0]]

X += xs
Y += ys
L1 += ["diff day rank 1"]
L2 += ["long-term"]


xs, ys = get_curve_ntrack(df_dict, "test_same_hour", rank=1)

X += xs
Y += ys
L1 += ["same hour rank 1"] * len(xs)
L2 += ["short-term"] * len(xs)


xs, ys = get_curve_ntrack(df_dict, "test_different_day_same_hour", rank=1)

X += xs
Y += ys
L1 += ["same hour diff day rank 1"] * len(xs)
L2 += ["short-term"] * len(xs)



xs, ys = get_curve_ntrack(df_dict, "test_different_day", rank=1)

X += xs
Y += ys
L1 += ["diff day rank 1"] * len(xs)
L2 += ["short-term"] * len(xs)


ddff = pd.DataFrame({"x":X, "y":Y, "eval": L1, "train":L2})
```
